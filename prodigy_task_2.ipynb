```python
#feature selection:
cols=[ 'Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch',
       'Fare', 'Embarked']
```

```python
X=train[cols]
Y=train['Survived']
```

```python
X['Age'].isnull().sum()
X['Pclass'].isnull().sum()

X['SibSp'].isnull().sum()
X['Parch'].isnull().sum()
X['Fare'].isnull().sum()
X['Sex'].isnull().sum()
X['Embarked'].isnull().sum()
```

```python
X['Age']=X['Age'].fillna(X['Age'].median())
X['Age'].isnull().sum()
```

```python
X['Embarked'] = train['Embarked'].fillna(method ='pad')
X['Embarked'].isnull().sum()
```

```python
d = {'male': 0, 'female': 1}
X['Sex'] = X['Sex'].apply(lambda x: d[x])
X['Sex'].head()
```

```python
e = {'C': 0, 'Q': 1, 'S': 2}
X['Embarked'] = X['Embarked'].apply(lambda x: e[x])
print(X['Embarked'].head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=7)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy=accuracy_score(Y_test,y_pred)
accuracy
```

```python
report=classification_report(Y_test,y_pred)
report
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
print(cm)
```

```python
from sklearn.svm import SVC
model1=SVC()
model1.fit(X_train,Y_train)
y_pred1=model1.predict(X_test)
```

```python
accuracy1=accuracy_score(y_pred1,Y_test)
accuracy1
```

```python
report1=classification_report(y_pred1,Y_test)
report1
```

```python
cm1= confusion_matrix(Y_test, y_pred1)
print(cm1)
```

```python
from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier()
model2.fit(X_train, Y_train)
y_pred2 = model2.predict(X_test)
```

```python
accuracy2=accuracy_score(y_pred2,Y_test)
accuracy2
```

```python
report2=classification_report(y_pred2,Y_test)
report2
```

```python
cm2= confusion_matrix(Y_test, y_pred2)
print(cm2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3=KNeighborsClassifier(n_neighbors=5)
model3.fit(X_train,Y_train)
y_pred3=model3.predict(X_test)
```

```python
accuracy3=accuracy_score(y_pred3,Y_test)
accuracy3
```
